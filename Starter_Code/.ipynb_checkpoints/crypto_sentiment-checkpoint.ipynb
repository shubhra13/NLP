```python
import os
import pandas as pd
%matplotlib inline
import nltk
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download/Update the VADER Lexicon
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()
```

# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Read your api key environment variable
news_api_key = os.getenv('NEWS_API')
```

```python
# Create a newsapi client
from newsapi.newsapi_client import NewsApiClient
news_api = NewsApiClient(news_api_key)
```

```python
from datetime import date, datetime, timedelta

def get_news(keyword, from_date=date.today(), daysback=30):
    """ Fetches x(default=30days) mo. news from news_api about a specific keyword.
       
       Args:
        keyword (str): keyword to search in the newsapi
        from_date (date, optional): Defaults to current date if not provided.
        daysback (int, optional): Defaults to 30 days if not provided.
    

    Returns:
        all_headlines (list): of type strings.
    """
      
    all_articles = [] 
    print(f"Fetching news about '{keyword}'")
    print("-"*30)
    past_date = from_date - timedelta(days=daysback)
    while from_date > past_date:
        print(f"retrieving news from: {from_date}")
        articles = news_api.get_everything(
            q=keyword,
            from_param=str(from_date),
            to=str(from_date),
            language="en",
            sort_by="relevancy",
            page=1,
        )
   
        text = []
        for i in range(0, len(articles["articles"])):
            all_articles.append(articles["articles"][i]) 
#         all_articles.append(text)
        from_date = from_date - timedelta(days=1) # looping through news from today until x days ago 
    return all_articles
```

```python
# Fetch the Bitcoin news articles
bitcoin_news= get_news("Bitcoin",date.today(),10)
```

```python
# Fetch the Ethereum news articles

ethereum_news = get_news("Ethereum",date.today(),10)
```

```python
def make_dataframe(list_to_df):
    """ Converts a list to dataframe.
       
       Args:
        list_to_df (list): keyword to search in the newsapi
        
    Returns:
        df (dataframe): Dataframe with Vadar Sentiment analysis.
        
    """
    sentiments = []    
    for article in list_to_df:
        try:
            content = article["content"]
#             print(f"---{content}")
            sentiment = analyzer.polarity_scores(content)            
            compound = sentiment["compound"]           
            pos = sentiment["pos"]
            neu = sentiment["neu"]
            neg = sentiment["neg"]


            sentiments.append({
                
                "Compound": compound,                
                "Negative": neg,
                "Neutral": neu,
                "Positive": pos,
                "text": content,

            })

        except AttributeError:
            pass

    # Create DataFrame
    df = pd.DataFrame(sentiments)

    # Reorder DataFrame columns
    cols = ["Compound", "Negative", "Neutral", "Positive","text"]
    df = df[cols]
    return df
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_df = make_dataframe(bitcoin_news)
bitcoin_df.head(30)
```

```python
# Create the ethereum sentiment scores DataFrame
ethereum_df = make_dataframe(ethereum_news)
ethereum_df.head(30)
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Expand the default stopwords list if necessary

sw = set(stopwords.words('english'))
sw_addon = {'chars', 'via', 'getty','Getty Images'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
       
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)  # Remove the punctuation
    words = word_tokenize(re_clean) # Create a list of the words    
    lemmatizer = WordNetLemmatizer() # Instantiate the lemmatizer
    lem = [lemmatizer.lemmatize(word) for word in words] # Lemmatize Words into root words
    
    # Convert the words to lowercase
    # Remove the stop words
    tokens = [word.lower() for word in words if word.lower() not in sw.union(sw_addon)] 
    return tokens
```

```python
# Create a new tokens column for bitcoin
token_column = []
for index, rows in bitcoin_df.iterrows():
    token_column.append(tokenizer(rows['text']))
bitcoin_df['token'] = token_column    
bitcoin_df.head()
```

```python
# Create a new tokens column for ethereum
token_column.clear()
for index, rows in ethereum_df.iterrows():
    token_column.append(tokenizer(rows['text']))
ethereum_df['token'] = token_column    
ethereum_df.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
def bigram_counter(text, n=2): 
    """ Creates the n gram counter
    Args:
        text (str): Articles for Ngrams
        n (int): Number for ngram, default is bigram
    

    Returns:
        bigram_counts (str): String with ngrams and counter.
    """
    bigrams = ngrams(text, n)
    bigram_counts = Counter(bigrams)
    return bigram_counts
```

```python
# Generate the Bitcoin N-grams where N=2
bitcoin_all_tokens = []
for index, row in bitcoin_df.iterrows(): 
    bitcoin_all_tokens = bitcoin_all_tokens + row['token']
bigram_counter(bitcoin_all_tokens[:20])
```

```python
# Generate the Ethereum N-grams where N=2
ethereum_all_tokens = []
for index, row in ethereum_df.iterrows(): 
    ethereum_all_tokens = ethereum_all_tokens + row['token']
bigram_counter(ethereum_all_tokens[:20])
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
token_count(bitcoin_all_tokens)
```

```python
# Get the top 10 words for Ethereum
token_count(ethereum_all_tokens)
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
all_bitoin_text = ' '.join(bitcoin_df['text'])
```

```python
# Run the NER processor on all of the text
doc = nlp(all_bitoin_text)

# Add a title to the document
doc.user_data["title"] = "Bitcoin NER"
```

```python
# Render the visualization
displacy.render(doc, style='ent')
```

```python
# List all Entities
for ent in doc.ents:
    print(ent.text, ent.label_)
```
